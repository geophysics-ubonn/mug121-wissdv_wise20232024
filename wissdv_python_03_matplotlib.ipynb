````python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
````

````python
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
````

````python
fig = plt.figure() # Initialisierung des figure-Objects
# Plot-Befehle:
plt.plot(x, y, label="plt.plot()")

plt.plot(x, y, marker=".", linestyle="--", label="plt.plot(marker='.', ls='--')")
plt.scatter(x, y, label="plt.scatter()")

plt.xlim(); plt.ylim() # Achsenlimits
plt.xlabel("X"); plt.ylabel("Y") # Achsenbeschriftung
plt.grid() # Gitter
plt.title("Titel") # Titel
plt.legend(loc="best") # Legende

plt.show() # Abschließendes "show"
````

````python
fig, ax = plt.subplots() # Initialisierung des figure- UND Achsenobjekts
#axes = ax[] # ax ist Liste von Achsenobjekten
ax.plot(x, y, label="plt.plot()"); ax.scatter(x, y, label="plt.scatter()") # Plot-Befehl
plt.plot(x, y, marker="o", linestyle="-", label="plt.plot(marker='o', ls='-')")
plt.scatter(x, y, marker="o", linestyle="-", label="plt.scatter(marker='o', ls='-')")

ax.set_xlim(); ax.set_ylim() # Achsenlimits
ax.set_xlabel("X"); ax.set_ylabel("Y") # Achsenbeschriftung
ax.set_xticks(x); ax.set_yticks(y) # Achsenabschnitte umdefinieren
ax.grid() # Gitter
ax.set_title("Title"); plt.suptitle("Suptitle") # Titel pro Achse und pro figure 
ax.legend(loc="best") # Legende mit automatischer Positionierung am Plotinhalt

plt.show() # Abschluss
````

````python
fig, ax = plt.subplots(2, 3, figsize=(18,5)) # Initialisierung des figure- UND Achsenobjekts

# Initialisierung des ersten Plots im 2 x 3 Subplot-Gitter
ax0 = ax[0, 0] # ax ist Liste von Achsenobjekten, als Matrix angeordnet [1. Zeile, 1. Spalte]
ax0.plot(x, y, "--", label="plt.plot()"); ax0.scatter(x, y, label="plt.scatter()")
ax0.grid() # Gitter
ax0.legend() # Legende

# Initialisierung des zweiten Plots im 2 x 3 Subplot-Gitter
ax1 = ax[0, 1] # ax ist Liste von Achsenobjekten, als Matrix angeordnet [1. Zeile, 2. Spalte]
ax1.loglog(x, y, "r+-", label="ax.loglog()") # Plot-Befehl
ax1.grid(axis="x")
ax1.legend() # Legende

# Initialisierung des dritten Plots im 2 x 3 Subplot-Gitter
ax2 = ax[1, 0] # ax ist Liste von Achsenobjekten, als Matrix angeordnet [2. Zeile, 1. Spalte]
ax2.semilogx(x, y, label="ax.semilogx()") # Plot-Befehl
ax2.grid(which="both")
ax2.legend() # Legende

# Initialisierung des vierten Plots im 2 x 3 Subplot-Gitter
ax3 = ax[1, 1] # ax ist Liste von Achsenobjekten, als Matrix angeordnet [2. Zeile, 2. Spalte]
ax3.semilogy(x, y, label="ax.semilogy()") # Plot-Befehl
ax3.grid(which="both")
ax3.legend() # Legende

# Fünfter Plot im 2 x 3 Subplot-Gitter bleibt leer
# ax4 = ax[0, 2]

# Initialisierung des sechsten Plots im 2 x 3 Subplot-Gitter
ax5 = ax[1, 2] # ax ist Liste von Achsenobjekten, als Matrix angeordnet [2. Zeile, 2. Spalte]
colormap = ax5.imshow([x, y], cmap="rainbow")
plt.colorbar(colormap, ax=ax5)
ax5.grid(which="both")
# ax5.legend() # Legende
ax5.set_title("Matrix aus Aufgabe 0") # Titel

plt.suptitle("Plot - Figure")

# plt.savefig("test.jpg", bbox_inches="tight")

plt.show() # Abschluss
````
